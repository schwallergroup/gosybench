```python
# Load a graph
import pickle

paper = "ja074300t"
with open("../data/papers/" + paper + "/gt_graph.pickle", "rb") as f:
    g = pickle.load(f)


import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore")

    for n in g.nodes():
        smiles = g.nodes[n].get("attr", {}).get("smiles", None)
        if smiles is None:
            pot_smi = py2opsin(
                chemical_name=g.nodes[n]
                .get("attr", {})
                .get("substance_name", "")
                or n,
                output_format="SMILES",
            )
            if pot_smi:
                if "attr" not in g.nodes[n]:
                    g.nodes[n]["attr"] = {}
                g.nodes[n]["attr"]["smiles"] = pot_smi
                print(
                    f'{n:<40} {pot_smi:<40} {g.nodes[n].get("attr",{}).get("substance_name", "")}'
                )

        # name = g.nodes[n].get("attr",{}).get("substance_name", "")
        # print(f'{n:<40} {name:<40} {g.nodes[n].get("attr",{}).get("reference_key", "")}')
```
